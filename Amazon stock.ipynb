```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
%matplotlib inline

amz = pd.DataFrame()
amz = pd.read_csv('HistoricalQuotes.csv',parse_dates=['date'])
amz
```

```python
amz.dtypes
```

```python
amz.head()
```

```python
amz.close = amz.close.apply(lambda x:x.replace(',',''))
amz.close = amz.close.astype(float)

amz.volume = amz.volume.apply(lambda x:x.replace(',',''))
amz.volume = amz.volume.astype(float)

amz.open = amz.open.apply(lambda x:x.replace(',',''))
amz.open = amz.open.astype(float)
```

```python
amz = amz.sort_values(by='date')
#amz.set_index('date',inplace=True)
amz['close'].plot(figsize=(10,5))
plt.ylabel('Open')
plt.xlabel('Dates')
```

```python
#Graph represents stronger closing in September
```

```python
amz = amz.sort_values(by='date')
amz.set_index('date', inplace=True)
amz['open'].plot(figsize=(10,5))
plt.ylabel('Open')
plt.xlabel('Dates')
```

```python
#Graph represents stronger opening in September
```

```python
amz = amz.sort_values(by='date')
amz['volume'].plot(figsize=(10,5))
plt.ylabel('Volume')
plt.xlabel('Dates')
```

```python
#Graph shows more volume in late october and early november
```

```python
plt.figure(figsize=(10, 5))

amz = amz.sort_values(by='date')
plt.plot(amz['open'], color='green')
plt.plot(amz['close'], color='red')
plt.ylabel('Open')
plt.xlabel('Dates')
plt.plot(amz['open'], label='Open') 
plt.plot(amz['close'], label='Close') 
plt.legend(loc='upper right') 
plt.show()
```
